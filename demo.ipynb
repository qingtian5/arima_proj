```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 读取 Excel 文件，并将第一列转换为时间变量
df = pd.read_excel('/workspace/arima_proj/data/data_2005.xlsx', parse_dates=[0])

# 假设第一列是时间，其他列是变量
missing_counts = df[df.iloc[:, 0].dt.year == 2022].isnull().sum()

# 提取季度变量
quarterly_variables = [column for column, count in missing_counts.items() if count == 8]

# 存储预测结果
predictions = {}

for column in quarterly_variables:
    # 提取每季度的值
    df['year'] = df.iloc[:, 0].dt.year
    df['quarter'] = df.iloc[:, 0].dt.quarter
    quarterly_data = df[['year', 'quarter', column]].dropna().set_index(['year', 'quarter'])[column]

    # 使用 SARIMA 进行预测
    model = SARIMAX(quarterly_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 4))  # 季节性参数可以根据需要调整
    model_fit = model.fit()

    # 进行预测
    forecast = model_fit.forecast(steps=8)  # 预测 2024 和 2025 年的每个季度
    predictions[column] = forecast

print(predictions)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 读取 Excel 文件，并将第一列转换为时间变量
df = pd.read_excel('/workspace/arima_proj/data/data_2005.xlsx', parse_dates=[0])

# 假设第一列是时间，其他列是变量
missing_counts = df[df.iloc[:, 0].dt.year == 2022].isnull().sum()

# 提取月度变量
monthly_variables = [column for column, count in missing_counts.items() if count == 0 and column != "时间"]
print(monthly_variables)
# 存储预测结果
predictions = {}

for column in monthly_variables:
    # 提取月度数据
    df['year'] = df.iloc[:, 0].dt.year
    df['month'] = df.iloc[:, 0].dt.month
    monthly_data = df.set_index(df.iloc[:, 0])[column].dropna()

    # 使用 SARIMA 进行预测
    model = SARIMAX(monthly_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # 季节性周期设置为12
    model_fit = model.fit()

    # 进行预测
    forecast = model_fit.forecast(steps=24)  # 预测 2024 和 2025 年的每个月
    predictions[column] = forecast

print(predictions)
```
